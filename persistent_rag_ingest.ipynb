```python
from ingest import load_data
from sqlitesearch import TextSearchIndex
import time
```

```python
data = load_data()
```

```python
index = TextSearchIndex(
    text_fields=['question', 'answer'],
    keyword_fields=['category'],
    db_path='data/ds_qa.db'
)
```

```python
for doc in data:
    index.add(doc)
    print(f"""Added: {doc["question"][:50]}...""")
    time.sleep(0.5)

index.close()
print("Done. Index saved to ds_qa.db")
```
